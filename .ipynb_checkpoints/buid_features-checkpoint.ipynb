```python
import pandas as pd
```

## build test set 

```python
df = pd.read_csv('./data/raw_test_wc.csv')
df.shape
```

```python
pairs = [
    ('title_ratio_keyword', 'title_num_keyword','title_num_word'),
    ('ratio_uni_char', 'num_uni_char','num_char'),
    ('ratio_uni_word', 'num_uni_word','num_word'),
    ('ratio_punc', 'num_punc', 'num_char'),
    ('ratio_noun', 'num_noun', 'num_word'),
    ('ratio_verb', 'num_verb', 'num_word'),
    ('ratio_adj', 'num_adj', 'num_word'),
    ('ratio_adv', 'num_adv', 'num_word'),
    ('ratio_conj', 'num_conj', 'num_word'),
    ('ratio_numeral', 'num_numeral', 'num_word'),
    ('ratio_char_para', 'num_char', 'num_para'),
    ('ratio_word_para', 'num_word', 'num_para')
]

for (c,a,b) in pairs:
    df[c] = df[a]/df[b]
    
df.head()
```

```python
df.to_csv('./data/test_wc.csv', header = True, index = False)
```

## build training set 

```python
df = pd.read_csv('./data/raw_train_wc.csv')
doctype = df.doctype.copy()
```

```python
df.drop(columns = ['doctype'], inplace = True)
df.columns
```

```python
pairs = [
    ('title_ratio_keyword', 'title_num_keyword','title_num_word'),
    ('ratio_uni_char', 'num_uni_char','num_char'),
    ('ratio_uni_word', 'num_uni_word','num_word'),
    ('ratio_punc', 'num_punc', 'num_char'),
    ('ratio_noun', 'num_noun', 'num_word'),
    ('ratio_verb', 'num_verb', 'num_word'),
    ('ratio_adj', 'num_adj', 'num_word'),
    ('ratio_adv', 'num_adv', 'num_word'),
    ('ratio_conj', 'num_conj', 'num_word'),
    ('ratio_numeral', 'num_numeral', 'num_word'),
    ('ratio_char_para', 'num_char', 'num_para'),
    ('ratio_word_para', 'num_word', 'num_para')
]

for (c,a,b) in pairs:
    df[c] = df[a]/df[b]
    
df.head()
```

```python
df['doctype'] = doctype
df.columns
```

```python
print(df.shape)
```

```python
df.to_csv('./data/train_wc.csv', header = True, index = False)
```
